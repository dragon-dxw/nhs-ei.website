```python

# strip any attrs we wont need them
REMOVE_ATTRIBUTES = [
    'lang','language','onmouseover','onmouseout','script','style','font',
    'dir','face','size','color','style','class','width','height','hspace',
    'border','valign','align','background','bgcolor','text','link','vlink',
    'alink','cellpadding','cellspacing']

soup = BeautifulSoup(content, "lxml", exclude_encodings=True)

# iframes are sometimes inside another tag so lets move them out of any parent tags
iframes = soup.find_all('iframe')

IFRAME_POSSIBLE_PARENTS = ['p', 'div', 'span'] # '[document]' means leave it alone

for iframe in iframes:
    parent = iframe.previous_element
    if parent.name in IFRAME_POSSIBLE_PARENTS:
        # print(parent)
        parent.replaceWith(iframe)

for attribute in REMOVE_ATTRIBUTES:
    for tag in soup.find_all(attrs={attribute: True}):
        del tag[attribute]

soup = soup.find('body').findChildren(recursive=False)

blocks = []

block_value = ''
counter = 0

for tag in soup:

    counter += 1

    """
    the process here loops though each soup tag to discover the block type to use
    there's a table and iframe block to deal with if they exist
    """

    # print(tag.name)
    if not tag.name in TAGS_TO_BLOCKS:
        # print(block_value)
        # it's a simple text field so concat all text
        block_value += str(tag)
        # when we reach the end and somehing is in the 
        # block_value just output and clear
        # blocks.append({
        #     'type': 'text',
        #     'value': block_value
        # })
        # block_value = ''

    if tag.name == 'table':

        if len(block_value) > 0:
            blocks.append({
                'type': 'text',
                'value': block_value
            })
            block_value = ''
        blocks.append({
            'type': 'html',
            'value': str(tag)
        })

    if tag.name == 'iframe':

        if len(block_value) > 0:
            blocks.append({
                'type': 'text',
                'value': block_value
            })
            block_value = ''
        blocks.append({
            'type': 'html',
            'value': '<div class="core-custom"><div class="responsive-iframe">{}</div></div>'.format(str(tag))
        })

    if counter == len(soup) and len(block_value) > 0:
        # when we reach the end and something is in the 
        # block_value just output and clear

        blocks.append({
            'type': 'text',
            'value': block_value
        })
        block_value = ''

# return blocks

blocks


```
